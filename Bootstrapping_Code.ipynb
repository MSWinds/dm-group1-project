## Load Original Breast Cancer Dataset

We will create 20 new observations based on the original Brest Cancer Dataset using Bootstrap Resampling.

```python
original_df = pd.read_excel(file_path + 'Breast Cancer.xlsx')
original_df.head(10)
```

```python
# Set the number of new datasets to generate
num_datasets = 1

# Create a list to store the new datasets
new_datasets = []

# Perform bootstrap resampling to generate new datasets
for i in range(num_datasets):
    # Sample observations with replacement from the existing dataset
    new_df = original_df.sample(n=20, replace=True)
    
    # Add the new dataset to the list
    new_datasets.append(new_df)
```

```python
# Concatenate the new datasets into a single DataFrame
new_df = pd.concat(new_datasets)
new_df.head(20)
```
