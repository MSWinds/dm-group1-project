### MERGE all the Data Frames

Merge all dataframes to create 1 data frame with following columns:
Gender, Country, Marital_Status, Number_of_Months, Filed_for_Bankrupcy, LongTermPlanningScore, DecisionStyle, PowerOrientedScore, CommunityOrientedScore

```python
# Compile all data frames
data_frames = [fam_income_df, parents_ed_df, credit_rating_df, loan_amount_df, HS_Academics_df]
```

```python
df_last_8 = reduce(lambda left, right: pd.merge(left, right, on = ['StudentID'], how = 'outer'), data_frames)
df_last_8.head(10)
```
