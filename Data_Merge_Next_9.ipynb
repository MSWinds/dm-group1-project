### Male Data

```python
# Read in Male.pl data file as a string
with open(file_path + 'male.pl', 'r') as f:
    male_data = f.read()

print(male_data)
```

```python
# Use regular expressions to parse the data into a list of dictionaries
matches = re.findall(r'\((.*?)\)', male_data)
male_data_list = []
for match in matches:
    split_match = match.split(',')
    data_dict = {'StudentID': split_match[0]}
    male_data_list.append(data_dict)

print(male_data_list[ :5])
```

```python
# Load the list of dictionaries into a pandas dataframe
male_df = pd.DataFrame(male_data_list)

male_df.head()
```

```python
# Remove 'student' so that only their ID shows only and then also add "Male" column.
male_df['StudentID'] = male_df['StudentID'].str.replace('student', '')
male_df['Gender'] = ('Male')
male_df.head(10)
```

```python
# Ensure that all StudentIDs in all the datasets are INT
male_df[['StudentID']] = male_df[['StudentID']].astype('int')

male_df.dtypes
```

### MERGE all the Data Frames

Merge all dataframes to create 1 data frame with following columns:
Gender, Country, Marital_Status, Number_of_Months, Filed_for_Bankrupcy, LongTermPlanningScore, DecisionStyle, PowerOrientedScore, CommunityOrientedScore

```python
# Compile all data frames
data_frames = [personality_df, male_df, absense_df, bankrupcy_df, c_ms_df]
```

```python
df_next_8 = reduce(lambda left, right: pd.merge(left, right, on = ['StudentID'], how = 'outer'), data_frames)
df_next_8.head(10)
```

```python
# For Gender, replace NaN's with Female
# For Filed_for_Bankrupcy, replace NaN's with 0; YES = 1, NO = 0
df_next_8['Gender'].fillna('Female', inplace=True)
df_next_8['Filed_for_Bankrupcy'].fillna('No', inplace = True)
df_next_8.head(10)
```
